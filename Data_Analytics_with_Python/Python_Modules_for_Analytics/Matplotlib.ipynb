# Data Visualization with Matplotlib

```python
import numpy as np
import pandas as pd
%matplotlib inline
import matplotlib.pyplot as plt
```

## Creating figures & axes

```python
# Setting classic stype
plt.style.use('seaborn-whitegrid') # you can to: classic

# start by creating figures and axes
fig = plt.figure()
ax = plt.axes()
```

## Subplots

```python
x = range(10)
y = range(10)
fig, ax = plt.subplots(nrows = 2, ncols = 3,
                      sharex = True, sharey = True
                      )
for row in ax:
    for col in row:
        col.plot(x, y)

plt.show()
```

## Wave Functions

```python
x = np.linspace(0, 10, 100)

# simple sinusoid graph
plt.plot(x, np.sin(x), '-')

# simple cosine graph
plt.plot(x, np.cos(x), '--');
```

```python
plt.figure() # create a plot figure

# create the first of two panels and set current axis
plt.subplot(2, 1, 1) # (rows, columns, panel number)
plt.plot(x, np.sin(x))

# create the second panel and set current axis
plt.subplot(2, 1, 2)
plt.plot(x, np.cos(x));
```

## Adjusting the Plot: Line colors & style

```python
plt.plot(x, np.sin(x - 0), color='blue') # specify color by name
plt.plot(x, np.sin(x - 1), color='g') # short color code (rgbcmyk)
plt.plot(x, np.sin(x - 2), color='0.75') # Grayscale between 0 and 1
plt.plot(x, np.sin(x - 3), color='#FFDD44') # Hex code (RRGGBB from 00 to FF)
plt.plot(x, np.sin(x - 4), color=(1.0,0.2,0.3)) # RGB tuple, values 0 and 1
plt.plot(x, np.sin(x - 5), color='chartreuse'); # all HTML color names supported
```

```python
plt.plot(x, x + 0, linestyle='solid')
plt.plot(x, x + 1, linestyle='dashed')
plt.plot(x, x + 2, linestyle='dashdot')
plt.plot(x, x + 3, linestyle='dotted');

# For short, you can use the following codes:
plt.plot(x, x + 4, linestyle='-') # solid
plt.plot(x, x + 5, linestyle='--') # dashed
plt.plot(x, x + 6, linestyle='-.') # dashdot
plt.plot(x, x + 7, linestyle=':'); # dotted
```

```python
# these linestyle and color codes can be combined into a single nonkeyword argument to the plt.plot() function
plt.plot(x, x + 0, '-g') # solid green
plt.plot(x, x + 1, '--c') # dashed cyan
plt.plot(x, x + 2, '-.k') # dashdot black
plt.plot(x, x + 3, ':r'); # dotted red
```

## Adjusting the Plot: Axes Limits

```python
plt.plot(x, np.sin(x))
plt.xlim(-1, 11)
plt.ylim(-1.5, 1.5)
```

## Labeling Plots

```python
plt.plot(x, np.sin(x))
plt.title("A Sine Curve")
plt.xlabel("x")
plt.ylabel("sin(x)");
```

```python
plt.plot(x, np.sin(x), '-g', label='sin(x)')
plt.plot(x, np.cos(x), ':b', label='cos(x)')
plt.axis('equal')
plt.legend();
```

## Customizing Legends

```python
x = np.linspace(0, 10, 1000)
fig, ax = plt.subplots()
ax.plot(x, np.sin(x), '-b', label='Sine')
ax.plot(x, np.cos(x), '--r', label='Cosine')
ax.axis('equal')
leg = ax.legend();
```

```python
ax.legend(loc='upper left', frameon=False)
fig
```

```python
ax.legend(frameon=False, loc='lower center', ncol=2)
fig
```

```python
ax.legend(fancybox=True, framealpha=1, shadow=True, borderpad=1)
fig
```

## Scatter Plots

```python
x = np.linspace(0, 10, 30)
y = np.sin(x)
plt.plot(x, y, 'o', color='black');
```

```python
rng = np.random.RandomState(0)
for marker in ['o', '.', ',', 'x', '+', 'v', '^', '<', '>', 's', 'd']:
    plt.plot(rng.rand(5), rng.rand(5), marker, label="marker='{0}'".format(marker))
    plt.legend(numpoints=1)
    plt.xlim(0, 1.8);
```

```python
plt.plot(x, y, '-ok'); # line (-), circle marker (o), black (k)
```

```python
plt.plot(x, y, '-p', color='gray', markersize=15, linewidth=4,markerfacecolor='white',markeredgecolor='gray',markeredgewidth=2)
plt.ylim(-1.2, 1.2);
```

```python
rng = np.random.RandomState(0)
x = rng.randn(100)
y = rng.randn(100)
colors = rng.rand(100)
sizes = 1000 * rng.rand(100)
plt.scatter(x, y, c=colors, s=sizes, alpha=0.5,cmap='viridis')
plt.colorbar(); # show color scale
```

```python
from sklearn.datasets import load_iris
iris = load_iris()
features = iris.data.T

plt.scatter(features[0], features[1], alpha=0.4,s=100*features[3], c=iris.target, cmap='viridis')
plt.xlabel(iris.feature_names[0])
plt.ylabel(iris.feature_names[1]);
```

## Error Bars

```python
x = np.linspace(0, 10, 50)
dy = 0.8
y = np.sin(x) + dy * np.random.randn(50)
plt.errorbar(x, y, yerr=dy, fmt='.k');
```

## Histograms, Binnings & Density

```python
data = np.random.randn(1000)
plt.hist(data, color='c', alpha=0.8)
```

```python
# The hist()function has many options to tune both the calculation and the display;
plt.hist(data, bins=30, alpha = 0.5, histtype='stepfilled', color='steelblue', edgecolor='none')
```

```python
x1 = np.random.normal(0, 0.8, 1000)
x2 = np.random.normal(-2, 1, 1000)
x3 = np.random.normal(3, 2, 1000)
kwargs = dict(histtype='stepfilled', alpha=0.3, bins=40)
plt.hist(x1, **kwargs)
plt.hist(x2, **kwargs)
plt.hist(x3, **kwargs);
```
